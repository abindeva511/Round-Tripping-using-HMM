```python
import numpy as np


n = 15  # size of the transition probability matrix

# Create an n x n transition probability matrix with random numbers
transition_matrix = np.random.rand(n, n)

# Set diagonal elements to zero
np.fill_diagonal(transition_matrix, 0)

# Normalize each row so that the sum of the row elements is 1
transition_matrix = transition_matrix / transition_matrix.sum(axis=1, keepdims=True)


```

```python


def metropolis_hastings(p, iterations, start_state, transition_matrix):
    """
    Metropolis-Hastings algorithm for Markov chain Monte Carlo (MCMC).
    p: Number of states
    iterations: Number of iterations to run the MCMC
    start_state: Initial state of the chain
    transition_matrix: Transition probability matrix
    """
    # Initialize the chain
    chain = [start_state]
    current_state = start_state
    
    for i in range(iterations):
        # Propose a new state from the transition matrix of the current state
        proposed_state = np.random.choice(p, p=transition_matrix[current_state])
        
        # Calculate the acceptance probability
        # Assuming a symmetric proposal distribution
        
        proposed_prob = transition_matrix[current_state][proposed_state]
        
        
        # Accept or reject the proposed state
        if np.random.uniform() < proposed_prob:
    
            current_state = proposed_state
        
            chain.append(current_state)
    
    return chain

# Example usage
p = 15  # Number of states
iterations = 3000  # Number of iterations for the MCMC
start_state = 0  # Starting state

# Transition matrix (assumed known for this example)


# Run MCMC
chain = metropolis_hastings(p, iterations, start_state, transition_matrix)


chain = [x + 1 for x in chain]

# Print the result
print("Generated chain:", chain)
```

# finding the clockwise or anticlockwise

```python
def calculate_signed_area(route):
    total_area = 0
    for i in range(1,len(route) - 2):
        city1 = route[i]
        city2 = route[i + 1]
        
        if city1 == "<" or city1 == ">":
            continue
        
        lat1, lon1 = cities_coor[city1]
        lat2, lon2 = cities_coor[city2]
        
        total_area += (lon2 - lon1) * (lat2 + lat1)
    
    return total_area / 2

l = []
for route in data:
    area = calculate_signed_area(route)
    direction = 1 if area > 0 else 0
    l.append(direction)
```

```python
route = [(y, i + 1,l[i]) for i, x in enumerate(data) for y in x]
```

```python
# HIDDEN STATE CREATING
# MAKE CONDITION FOR THE FUTURE
# t


hidd = {}
for i in range(len(route)):
  hidd[i] = []
  for j in range(len(data)):
    if route[i][0] in data[j] and route[i][2] == l[j]:
      hidd[i].append(j)

route = [(y, i + 1) for i, x in enumerate(data) for y in x]

# hidden states
hidd_unique = sorted(list(set(tuple(x) if isinstance(x, list) else x for x in hidd.values())))
hidd_encod = {}
for i in range(len(hidd_unique)):
  hidd_encod[f"hidden_{i}"] = hidd_unique[i]



train = []
for i in range(len(route)):
  for j in range(len(hidd_unique)):
    if list(hidd_encod[f"hidden_{j}"]) == hidd[i]:
      train.append(str(route[i][0]) + "\t" + f"hidden_{j}" + "\n")


with open("round_trip_routes.pos", "w") as file:
    file.writelines(train)
```
